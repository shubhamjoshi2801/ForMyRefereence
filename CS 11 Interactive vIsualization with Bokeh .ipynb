```python
from bokeh.plotting import figure,show
from bokeh.io import output_notebook
output_notebook()
```

```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
sales=pd.read_excel('storesales.xlsx')
```

```python
sales['Order Date']=pd.to_datetime(sales['Order Date'])
sales['Order Month']=sales['Order Date'].dt.month.values
sales['Order Day']=sales['Order Date'].dt.day.values
sales['Order Year']=sales['Order Date'].dt.year.values
```

```python
sales.set_index('Order Date',inplace=True)
```

```python
#find the monthly frequency of average sales for all the years:
```

```python
avg_sales_variation=figure(plot_width=450,plot_height=450)
avg_sales_variation.line(x=sales['2009':'2012']['Sales'].resample('W').mean().index,y=sales['2009':'2012']['Sales'].resample('W').mean().values,color='red')
show(avg_sales_variation)
```

```python
max_sales=pd.DataFrame()
for year,df_year in sales.groupby('Order Year'):
    for month,df_month in df_year.groupby('Order Month'):
        max_sales=max_sales.append(df_month.sort_values(by='Sales',ascending=False).iloc[0,:])
        
from bokeh.plotting import figure,show
from bokeh.io import output_notebook
output_notebook()

max_sales_fig=figure(plot_width=450,plot_height=450)
years=[2009,2010,2011,2012]
colors=['green','blue','orange','red']
for count in range(len(years)):
    max_sales_fig.line(x=max_sales[max_sales['Order Year']==years[count]]['Order Month'],
                      y=max_sales[max_sales['Order Year']==years[count]]['Sales'],color=colors[count],line_width=3,alpha=0.4,
                      legend=str(years[count]))
    
    max_sales_fig.scatter(x=max_sales[max_sales['Order Year']==years[count]]['Order Month'],
                      y=max_sales[max_sales['Order Year']==years[count]]['Sales'],color=colors[count],size=4)
show(max_sales_fig)    
```

```python
max_sales=pd.DataFrame()
for year,df_year in sales.groupby('Order Year'):
    for month,df_month in df_year.groupby('Order Month'):
        max_sales=max_sales.append(df_month.sort_values(by='Sales',ascending=False).iloc[0,:])
        
from bokeh.plotting import figure,show
from bokeh.io import output_notebook
output_notebook()

max_sales_fig=figure(plot_width=450,plot_height=450)
for count in range(len(years)):
    max_sales_fig.vbar(x=max_sales[max_sales['Order Year']==2009]['Order Month'],
                      top=max_sales[max_sales['Order Year']==2009]['Sales'],color='blue',width=0.6)
    
show(max_sales_fig)    
```

```python
max_sales=pd.DataFrame()
for year,df_year in sales.groupby('Order Year'):
    for month,df_month in df_year.groupby('Order Month'):
        max_sales=max_sales.append(df_month.sort_values(by='Sales',ascending=False).iloc[0,:])
        
from bokeh.plotting import figure,show
from bokeh.io import output_notebook
output_notebook()

max_sales_fig=figure(plot_width=450,plot_height=450)
for count in range(len(years)):
    max_sales_fig.line(x=max_sales[max_sales['Order Year']==2009]['Order Month'],
                      y=max_sales[max_sales['Order Year']==2009]['Sales'],color='blue',legend='2009')
    
show(max_sales_fig)    
```

```python
#plot the annual variation of average monthly sales:
avg_monthly_sales=[]
years=[]
months=[]
for year,df_year in sales.groupby('Order Year'):
    for month,df_month in df_year.groupby('Order Month'):
        avg_monthly_sales.append(df_month['Sales'].mean())
        years.append(year)
        months.append(month)
avg_month_df=pd.DataFrame(data={'Year':years,'Month':months,'Avg_Sales':avg_monthly_sales})
```

```python
from bokeh.plotting import figure,show
from bokeh.io import output_notebook
output_notebook()
```

```python
avg_monthly_sales=figure(plot_width=500,plot_height=500)
avg_monthly_sales.line(x=avg_month_df[avg_month_df['Year']==2012]['Month'],
                       y=avg_month_df[avg_month_df['Year']==2012]['Avg_Sales'],
                       line_width=1.75,legend='2009',color='red')
avg_monthly_sales.title.text='Annual variation of avergae monthly sales'
avg_monthly_sales.xaxis.axis_label='month'
avg_monthly_sales.yaxis.axis_label='avg_sales'
show(avg_monthly_sales)
```
