```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
sales=pd.read_excel('storesales.xlsx')
```

```python
sales['Order Date']=pd.to_datetime(sales['Order Date'])
sales['Order Month']=sales['Order Date'].dt.month.values
sales['Order Day']=sales['Order Date'].dt.day.values
sales['Order Year']=sales['Order Date'].dt.year.values
sales.set_index('Order Date',inplace=True)
```

#### 1)Without using the 'groupby' as much as method, plot the weekly,monthly and average variation of total sales for every years :

```python
from bokeh.plotting import figure,show
from bokeh.io import output_notebook
output_notebook

def periodic_value_avg(dataframe,year,freq):
        frequency=dataframe.loc[str(year),'Sales'].resample(freq).mean().index
        frequency_avg=dataframe.loc[str(year),'Sales'].resample(freq).mean().values
        periodic_avg=figure(plot_width=450,plot_height=450)
        if freq =='W':
            periodic_avg.line(x=frequency.week.values,y=frequency_avg,legend=str(year),line_width=1.75)
            periodic_avg.xaxis.axis_label='weeks'
            periodic_avg.yaxis.axis_label='avg_weekly_sales'
        
        if freq =='M':
            periodic_avg.line(x=frequency.month.values,y=frequency_avg,legend=str(year),line_width=1.75)
            periodic_avg.xaxis.axis_label='months'
            periodic_avg.yaxis.axis_label='avg_monthly_sales'
            
        return show(periodic_avg)
        
periodic_value_avg(sales,2009,freq='M')
```

#### 5) Find the monthly total of furniture and technology good shipped to ontario and brtisih colombia through express air and regular air.

```python
list_province=['Ontario','British Columbia']
list_delivery=['Express Air','Regular Air']
list_category=['Technology','Furniture']
sales_final=sales[(sales['Province'].isin(list_province))
                    &(sales['Product Category'].isin(list_category))
                    &(sales['Ship Mode'].isin(list_delivery))]
sales_final.loc['2009':'2012','Sales'].resample('M').sum().head()
```

```python
from bokeh.plotting import figure,show
from bokeh.io import output_notebook
output_notebook()
```

```python
monthly_total=figure(plot_width=500,plot_height=500)

monthly_total.xaxis.axis_label='months'
monthly_total.yaxis.axis_label='total_sales'

years=['2009','2010','2011','2012']
colors=['red','green','orange','blue']

for count in range(len(years)):
    X=sales_final.loc[years[count],'Sales'].resample('M').sum().index.month
    Y=sales_final.loc[years[count],'Sales'].resample('M').sum().values
    monthly_total.line(x=X,y=Y,color=colors[count],legend=years[count],line_width=1.75)
    monthly_total.scatter(x=X,y=Y,color=colors[count],legend=years[count])


show(monthly_total)
```
